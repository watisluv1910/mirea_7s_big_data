## Задание №6

Используя признак charges или imb, проверить, выполняется ли
центральная предельная теорема. Использовать различные длины выборок
n. Количество выборок = 300. Вывести результат в виде гистограмм. Найти
стандартное отклонение и среднее для полученных распределений.
Сделать выводы.

**Теория**:

Центральная предельная теорема (ЦПТ) утверждает, что независимо от того, как распределена исходная выборка, распределение выборочных средних (если брать достаточно много выборок) стремится к нормальному распределению по мере увеличения размера выборки `n`. Это означает, что даже если данные исходно имеют не нормальное распределение, средние по выборкам будут распределяться по нормальному закону, если `n` достаточно велико.

**ЦПТ** лежит в основе многих статистических методов. Благодаря ей можно использовать нормальное распределение для оценки средних значений, даже если сами данные не подчиняются нормальному распределению.

````python
# Центральная предельная теорема (используем charges)
sample_sizes = [10, 30, 50, 100]  # Различные длины выборок
num_samples = 300  # Количество выборок

plt.figure(figsize=(14, 10))
for i, n in enumerate(sample_sizes, 1):
    sample_means = [np.mean(np.random.choice(df['charges'], size=n)) for _ in range(num_samples)]
    plt.subplot(2, 2, i)
    plt.hist(sample_means, bins=20, color='lightblue', edgecolor='black')
    plt.title(f'Гистограмма средних выборок (n={n})')
    plt.xlabel('Среднее значение charges')
    plt.ylabel('Частота')

plt.tight_layout()
plt.show()

# Среднее и стандартное отклонение для полученных распределений (n = 50)
sample_means_50 = [np.mean(np.random.choice(df['charges'], size=50)) for _ in range(num_samples)]
mean_50 = np.mean(sample_means_50)
std_50 = np.std(sample_means_50)

mean_50, std_50
````

Были использованы данные о расходах на медицинскую страховку (charges), для которых были вычислены средние значения для 300 выборок разной длины `n = 10, 30, 50, 100`. По гистограммам видно, что с увеличением `n`, распределение средних становится все более нормальным, что подтверждает **ЦПТ**.

**Вывод**:
- Для малых выборок (например, `n = 10`) распределение средних еще довольно «размытое» и отдалено от нормального;
- С увеличением длины выборки (например, `n = 100`) распределение средних становится почти нормальным.

Это подтверждает, что **ЦПТ** работает и можно полагаться на нормальные методы для анализа средних значений.

## Задание №7

Построить 95% и 99% доверительный интервал для среднего значения
расходов и среднего значения индекса массы тела.

**Теория**:

Доверительный интервал показывает диапазон значений, в котором с определенной вероятностью находится истинное среднее значение. Например, для 95%-го доверительного интервала можно быть уверенным, что в 95% случаев истинное среднее попадет в этот диапазон.

````python
# Доверительный интервал для charges и bmi
confidence_intervals = {}

for col in ['charges', 'bmi']:
    mean = np.mean(df[col])
    std_err = stats.sem(df[col])
    confidence_95 = stats.t.interval(0.95, len(df[col])-1, loc=mean, scale=std_err)
    confidence_99 = stats.t.interval(0.99, len(df[col])-1, loc=mean, scale=std_err)
    confidence_intervals[col] = {'95% CI': confidence_95, '99% CI': confidence_99}
    
# Вывод доверительных интервалов
for col, intervals in confidence_intervals.items():
    print(f"Для признака '{col}':")
    print(f"  95% доверительный интервал: {intervals['95% CI'][0]:.2f} - {intervals['95% CI'][1]:.2f}")
    print(f"  99% доверительный интервал: {intervals['99% CI'][0]:.2f} - {intervals['99% CI'][1]:.2f}")
````

Для расходов на медицинскую страховку с вероятностью 95% средние расходы на страховку лежат в диапазоне от 12,621 до 13,920 долларов, а с вероятностью 99% — в диапазоне от 12,416 до 14,124 долларов.

Для индекса массы тела (BMI) с вероятностью 95% средний индекс массы тела лежит в диапазоне от 30.34 до 30.99, а с вероятностью 99% — в диапазоне от 30.23 до 31.09.

**Вывод**:

Интервалы показывают разумную уверенность в истинных средних значениях. При этом для расходов доверительные интервалы шире, так как данные более разбросаны.

## Задание №8

Проверить распределения следующих признаков на нормальность: индекс
массы тела, расходы. Сформулировать нулевую и альтернативную
гипотезы. Для каждого признака использовать KS-тест и q-q plot. Сделать
выводы на основе полученных p-значений.

**Теория**:

Нормальное (гауссово) распределение — это одно из самых распространённых и важных распределений вероятностей в статистике.

Основные характеристики нормального распределения:

1.	Форма колокола: Нормальное распределение симметрично относительно своего среднего значения, и его график напоминает форму колокола. Это значит, что большая часть значений сосредоточена около среднего, и по мере удаления от него вероятность значений уменьшается;
2.	Среднее, мода и медиана совпадают: В нормальном распределении среднее значение, медиана (центральное значение) и мода (наиболее часто встречающееся значение) находятся в одной точке — посередине колоколообразной кривой;
3.	Стандартное отклонение (σ): Стандартное отклонение определяет, насколько широко распределены данные вокруг среднего значения. При большом стандартном отклонении распределение становится более плоским и растянутым, а при малом — более узким и высоким.

Проверка нормальности — это метод для определения того, насколько данные следуют нормальному распределению.

1. KS-тест (Критерий Колмогорова-Смирнова) - проверяет гипотезу о том, что данные взяты из некоторого распределения (например, нормального). В KS-тесте используются следующие гипотезы:
    - Нулевая гипотеза: Данные следуют нормальному распределению.
    - Альтернативная гипотеза: Данные не следуют нормальному распределению
    
    Если p-значение меньше выбранного уровня значимости (обычно 0.05), то нулевая гипотеза отвергается, и можно сделать вывод, что данные не распределены нормально;

2. Q-Q Plot (Квантиль-квантильный график) — это графический метод для проверки нормальности. На этом графике сравниваются квантили данных с квантилями нормального распределения. Если точки на графике располагаются вдоль диагональной линии, это говорит о том, что данные следуют нормальному распределению.

````python
# Проверка нормальности с использованием KS-теста и Q-Q plot
plt.figure(figsize=(14, 6))
for i, col in enumerate(['bmi', 'charges'], 1):
    plt.subplot(1, 2, i)
    stats.probplot(df[col], dist="norm", plot=plt)
    plt.title(f'Q-Q Plot для {col}')

plt.tight_layout()
plt.show()

# KS-тест для нормальности
ks_results = {}
for col in ['bmi', 'charges']:
    ks_stat, p_value = stats.kstest((df[col] - df[col].mean()) / df[col].std(), 'norm')
    ks_results[col] = {'KS-statistic': ks_stat, 'p-value': p_value}
    
# Вывод результатов KS-теста
for col, ks in ks_results.items():
    print(f"Распределение для признака '{col}':")
    print(f"  KS-статистика: {ks['KS-statistic']:.4f}")
    print(f"  P-значение: {ks['p-value']:.4g}")
    if ks['p-value'] < 0.05:
        print(f"  Вывод: Распределение не является нормальным (p < 0.05)")
    else:
        print(f"  Вывод: Распределение является нормальным (p > 0.05)")
    print()
````

Индекс массы тела (BMI):
- p-значение $0.316$, что значительно выше 0.05. Это означает, что мы не можем отвергнуть гипотезу о нормальности. Другими словами, BMI можно считать нормальным.
- Q-Q plot также подтверждает это, поскольку точки лежат близко к прямой.

Расходы на медицинскую страховку (charges):
- p-значение $4.38 \times 10^{-42}$, что значительно ниже 0.05. Это означает, что распределение расходов не является нормальным.
- Q-Q plot показывает отклонения от прямой, что указывает на ненормальность распределения.

**Выводы**:

Индекс массы тела можно считать нормально распределенным. Расходы на медицинскую страховку не подчиняются нормальному распределению, что видно как из теста, так и графически.

Загрузить данные из файла “bmi.csv”. Взять оттуда две выборки. Одна выборка – это индекс массы тела людей c региона northwest, вторая выборка – это индекс массы тела людей с региона southwest. Сравнить средние значения этих выборок, используя t-критерий Стьюдента. Предварительно проверить выборки на нормальность (критерий Шопиро-Уилка) и на гомогенность дисперсии (критерий Бартлетта).

**Теория**:

1.	Критерий Шапиро-Уилка используется для проверки нормальности распределения. Нулевая гипотеза заключается в том, что данные распределены нормально. Если $p \text{-value} > 0.05$, гипотеза о нормальности не отвергается.
2.	Критерий Бартлетта проверяет гомогенность дисперсий, то есть равенство дисперсий нескольких выборок. Нулевая гипотеза утверждает, что дисперсии выборок равны. Если $p \text{-value} > 0.05$, дисперсии можно считать одинаковыми.
3.	T-критерий Стьюдента используется для сравнения средних значений двух выборок. Нулевая гипотеза заключается в том, что средние значения выборок равны. Если $p \text{-value} < 0.05$, то средние значимо различаются.

Гомогенность дисперсий — это предположение о том, что дисперсии (разброс данных) двух или более выборок одинаковы.

````python
bmi_data = pd.read_csv('./data/pract3/bmi.csv')

# Выборки по регионам northwest и southwest
northwest_bmi = bmi_data[bmi_data['region'] == 'northwest']['bmi']
southwest_bmi = bmi_data[bmi_data['region'] == 'southwest']['bmi']

# Проверка выборок на нормальность с помощью критерия Шапиро-Уилка
shapiro_northwest = stats.shapiro(northwest_bmi)
shapiro_southwest = stats.shapiro(southwest_bmi)

# Проверка гомогенности дисперсий с помощью критерия Бартлетта
bartlett_test = stats.bartlett(northwest_bmi, southwest_bmi)

# Сравнение средних значений выборок с использованием t-критерия Стьюдента
t_test_result = stats.ttest_ind(northwest_bmi, southwest_bmi)

shapiro_northwest, shapiro_southwest, bartlett_test, t_test_result
````

**Выводы**:

1.	Проверка на нормальность (критерий Шапиро-Уилка):

	<br>Оба значения $p$-уровня больше 0.05, что позволяет принять нулевую гипотезу о нормальности распределений для обеих выборок. Следовательно, распределения индексов массы тела в этих выборках нормальны;

2.	Проверка гомогенности дисперсий (критерий Бартлетта) $p \text{-value} = 0.0652$. Это значение больше 0.05, что позволяет принять нулевую гипотезу о равенстве дисперсий (гомогенности). Следовательно, дисперсии этих выборок можно считать одинаковыми;
<br>
3.	Сравнение средних значений (t-критерий Стьюдента) $p \text{-value} = 0.0011$. Поскольку $p$-значение меньше 0.05, мы отвергаем нулевую гипотезу о равенстве средних значений индексов массы тела. ## Задание №14

Кубик бросили 600 раз, получили следующие результаты:

| N | Количество выпадений |
|---|----------------------|
| 1 | 97                   |
| 2 | 98                   |
| 3 | 109                  |
| 4 | 95                   |
| 5 | 97                   |
| 6 | 104                  |

С помощью критерия Хи-квадрат проверить, является ли полученное распределение равномерным. Использовать функцию *scipy.stats.chisquare()*.

**Теория**:

*Равномерное распределение* — это распределение, при котором все исходы случайного эксперимента имеют одинаковую вероятность. В контексте кубика это означает, что вероятность выпадения каждой грани равна $\frac{1}{6}$, и при большом количестве бросков все грани выпадают примерно одинаковое количество раз.

*Критерий Хи-квадрат* используется для проверки соответствия наблюдаемого распределения некоторому ожидаемому распределению. Нулевая гипотеза состоит в том, что наблюдаемое распределение не отличается от ожидаемого. Если $p$-значение меньше 0.05, нулевая гипотеза опровергается.

Функция *scipy.stats.chisquare()* вычисляет значение статистики $\chi^2$ и соответствующее $p$-значение, которые помогают оценить, насколько сильно наблюдаемое распределение отличается от ожидаемого.

````python
# Данные по количеству выпадений каждой грани кубика
observed_frequencies = [97, 98, 109, 95, 97, 104]

# Ожидаемые частоты для равномерного распределения (т.к. кубик честный, то вероятность выпадения каждой грани равна 1/6)
expected_frequencies = [600 / 6] * 6  # равномерное распределение, 600 бросков / 6 граней

stats.chisquare(observed_frequencies, expected_frequencies)
````

$\chi^2 \text{-статистика} = 1.44$
$p \text{-value} = 0.9199$

Поскольку $p$-значение значительно больше 0.05, нельзя опровергнуть нулевую гипотезу о равномерности распределения. Это означает, что результаты бросков кубика соответствуют равномерному распределению.
